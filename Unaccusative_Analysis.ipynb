```python
import re
import csv
```

```python
def create_wb():
    wb = {}
    lang_ids = {}
    with open('viborka.csv', 'r', encoding='utf-8') as csvfile:
        csvlines = csv.reader(csvfile, delimiter=',')
        start = 0
        for row in csvlines:
            if start == 0:
                for index, lang in enumerate(row):
                    wb[index] = []
                    lang_ids[index] = lang
                start += 1
            else:
                predicate = row[0]
                for index, block in enumerate(row):
                    res = block.strip()
                    if res == '1' or res == 1:
                        wb[index].append(predicate)
    tuple_wb = {}
    for key in wb:
        pr_list = wb[key]
        lang = lang_ids[key]
        if not re.match('[А-ЯЁ]', lang):
            for i in pr_list:
                for j in pr_list:
                    if i != j:
                        pair = tuple(sorted([i, j]))
                        if pair not in tuple_wb:
                            tuple_wb[pair] = [lang]
                        elif pair in tuple_wb and lang not in tuple_wb[pair]:
                            tuple_wb[pair].append(lang)
    for item in sorted(tuple_wb, key=lambda item: len(tuple_wb[item]), reverse=True):
        if len(tuple_wb[item]) >= 3:
            print(str(item) + '\t---\t' + ','.join(tuple_wb[item]) + '\t---\t' + str(len(tuple_wb[item])))
```

```python
create_wb()
```
